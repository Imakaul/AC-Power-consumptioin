```python
import pandas as pd 
import numpy as np 
df = pd.read_csv("AC_Data.csv")
df = df.rename(columns={'0':'Date'})
df['Date'] = pd.to_datetime(df['Date'])
df = df.dropna( how='all',subset=['AC 1'])
columns = ['AC 1', 'AC 2', 'AC 3', 'AC 4', 'AC 5', 'AC 6', 'AC 7',
       'AC 8', 'AC 9', 'AC 10', 'AC 11', 'AC 12', 'AC 13', 'AC 14', 'AC 15',
       'AC 16', 'AC 17', 'AC 18']
df[columns] = df[columns].fillna(df[columns].mean())
Date_wise = df.groupby([df.Date.dt.date]).sum()
Date_wise = Date_wise.reset_index()
```

# First we will clean the data and bring it into a better form so as we can apply visualisations on it. 

We have removed the NAN values from the dataset and we have put the mean values in the missing value of the respective columns. After that we have changed the Date coulmn in Date and time format. 

```python
Date_wise.head()
```

```python
Date_wise.sum()
```

As from the above sum of power consumption we can easily predict that "AC 18" was used the most and "AC 5" was used the least. 

```python
import plotly.express as px
fig = px.line(Date_wise, x="Date", y=Date_wise.columns,
              hover_data={"Date": "|%B %d, %Y"},
              title='Power Consumption')
fig.update_layout(title_x=0.5,template = 'simple_white',width=1000)
fig.update_xaxes(
    dtick="M1",
    tickformat="%b\n%Y")
fig.show()
```

```python
hour_wise = df.groupby([df.Date.dt.hour]).sum()
hour_wise = hour_wise.reset_index()
```

```python
hour_wise.head()
```

```python
fig = px.line(hour_wise, x="Date", y=hour_wise.columns,
              hover_data={"Date": "|%B %d, %Y"},
              title='Hour Wise Consumption')
fig.update_layout(title_x=0.5,template = 'simple_white',width=1000)
fig.update_xaxes(
    dtick="M1",
    tickformat="%b\n%Y")
fig.show()
```

By using plotly we are able to derive the relationship between AC power consumption per day for a period of 2 months. 

```python
import plotly.express as px
import plotly.graph_objects as go
fig = px.histogram(Date_wise, x="Date", y=Date_wise.columns, histfunc="avg", title="AC Consumption on daily basis")
fig.update_traces(xbins_size="M1")
fig.update_xaxes(showgrid=True, ticklabelmode="period", dtick="M1", tickformat="%b\n%Y")
fig.update_layout(bargap=0.1)
fig.add_trace(go.Scatter(mode="markers", x=Date_wise["Date"], y=Date_wise.columns, name="daily"))
fig.update_layout(title_x=0.5,template = 'simple_white',width=1000)

fig.show()
```

This histogram represents the avg power consumed by the AC's over a period of two months.

```python
fig = px.histogram(hour_wise, x="Date", y=hour_wise.columns, histfunc="avg", title="AC Consumption on Hourly basis")
fig.update_traces(xbins_size="M1")
fig.update_xaxes(showgrid=True, ticklabelmode="period", dtick="M1", tickformat="%b\n%Y")
fig.update_layout(bargap=0.1)
fig.add_trace(go.Scatter(mode="markers", x=hour_wise["Date"], y=hour_wise.columns, name="daily"))
fig.update_layout(title_x=0.5,template = 'simple_white',width=1000)

fig.show()
```

This Histogram represents the power consumption on Hourly basis for the period of two months.

```python
fig = px.box(Date_wise, x="Date", y=Date_wise.columns,title="2 Months consumption")
fig.update_layout(title_x=0.5,template = 'simple_white',width=1000)

fig.show()
```

This Box plot represents that how much power is consumed by all the ACs in a day over a period of 2 months. 

```python
fig = px.box(hour_wise, x="Date", y=hour_wise.columns,title="Hourly Consumption")
fig.update_layout(title_x=0.5,template = 'simple_white',width=1000)

fig.show()
```

This Box plot represents how much power is consumed by AC's on hourly basis.
